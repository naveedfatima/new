Droping Columns

```python
columns = ['Ticket', 'Cabin', 'Name']
```

```python
df = df.drop(columns, axis=1)
```

```python
df
```

Looking for null values

```python
df.isnull().sum()
```

Replacing nan value in Age by median

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df
```

replacing nan value in Embarked colummn by mode

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df
```

```python
df.isnull().sum()
```

Replacing female by "1" and male by "0" values in sex colummn

```python
df['Gender'] = df['Sex'].map({'female': 0, 'male': 1}).astype(int)
```

```python
df
```

Cmparing Age and fare graphically

```python
x=df['Age']
y=df['Fare']
plt.scatter(x, y, edgecolors='black', facecolor='yellow')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('scatter plot of Fare and Age')
plt.show()
```

```python
x=df['Age']
y=df['Fare']
colors=['red' if age >30 else 'green' for age in x]
plt.scatter(x, y, c=colors)
plt.xlabel('Age (red if > 30)')
plt.ylabel('Fare (green if <= 30)')
plt.title('scatter plot of Fare and Age')
plt.show()
```

```python
x = df['Age']
y = df['Fare']
def colors(col):
    if 1 <= col['Age'] <= 10:
        return 'green'
    elif 11 < col['Age'] <= 20:
        return 'yellow'
    elif 20 < col['Age'] <= 30:
        return 'orange'
    elif 30 < col['Age'] <= 40:
        return 'red'
    elif 40 < col['Age'] <= 50:
        return 'blue'
    elif 50 < col['Age'] <= 60:
        return 'black'
    elif 60 < col['Age'] <= 70:
        return 'pink'
    elif 70 < col['Age'] <= 80:
        return 'brown'
    elif 80 < col['Age'] <= 90:
        return 'purple'
    elif 90 < col['Age'] <= 100:
        return 'aqua'
    else:
        return 'navy'



```

```python
df['color'] = df.apply(colors, axis=1)
```

```python
plt.scatter(x, y, c=df['color'])
plt.title('Scatter Plot of Age vs Fare', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
